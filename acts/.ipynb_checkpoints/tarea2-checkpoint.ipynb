Creo lista que guarda(agregando con .append), para cada nombre, la suma de sus correspondientes evaluaciones 1 y 2, puesto que la linea i de nombres corresponde a la linea i de eval1 y eval2. La lista es de tuplas, en la que la posicion 0 es el nombre y la 1 la suma de las evaluaciones 1 y 2.

```python
lista_estudiantes = []
for i in range(len(nombres)):
    lista_estudiantes.append((nombres[i],eval1[i]+eval2[i]) 
```

Suma en una variable el total de las notas, a partir de recorrer la lista de estudiantes e ir tomando la posicion 1 de la segunda posicion de la tupla. Luego calcula el promedio dividiendo por el largo de la lista (cantidad de estudiantes).

```python
total_notas = 0
for i in range(len(lista_estudiantes)):
    total_notas += lista_estudiantes[i][1]
promedio = int(total_notas/len(lista_estudiantes))
```

Imprime las posiciones de la lista si, por cada estudiante, la posicion 1 de la tupla (segunda posicion porque empieza por 0) es menor al promedio.

```python
for i in range(len(lista_estudiantes)):
    if lista_estudiantes[i][1] < promedio:
        print(lista_estudiantes[i])
```
